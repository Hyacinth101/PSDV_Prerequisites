### Palindrome Check:
Write a program to check if a string is a palindrome (ignoring spaces and case).
Input: "A man a plan a canal Panama"
Output: True

```python
n= "A man a plan a canal Panama"
n= ''.join(n.split()).lower()
if n==n[::-1]:
    print ("True")
```

### Q.9) Merging Dictionaries:
###### Write Python code to merge two dictionaries. If a key is present in both, sum their values.

Example:

dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
Output: {'a': 10, 'b': 25, 'c': 15}

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}

for key, value in dict2.items():
    if key in dict1:
        dict1[key] += value  
    else:
        dict1[key] = value  


print(dict1)
```

### Q.10) Character Frequency:
Write a function that accepts a string and:
Converts all characters to lowercase.
Counts the frequency of each character.
Returns a dictionary with characters as keys and their counts as values.
Visualize the character frequencies using a bar chart.



```python
n= (input()).lower()
l={}
for word in n:
    if word in l:
        l[word]+=1
    else:
        l[word]=1
print (l)

import matplotlib.pyplot as plt

plt.bar(l.keys(),l.values())
plt.xlabel('CHARACTERS')
plt.ylabel(" FREQ")
plt.show()
```

## Remove Duplicates:
#### Write a program to remove duplicate elements from a list while maintaining the order of elements.
#### Calculate the mean and median of the unique elements.

```python

l = [1, 2, 2, 3, 4, 5, 5, 6,7,8,9,10]
k = {}
for word in l:
    if word in k:
        k[word] += 1
    else:
        k[word] = 1

m = [key for key, value in k.items() if value == 1]
print(m)

import statistics
mean = statistics.mean(m)
median = statistics.median(m)

print("Mean:", mean)
print("Median:", median)


```

### String Operations:
Given a list of strings, write a Python function to:
Find the longest string.
Count how many strings start with a vowel.
Visualize the distribution of string lengths using a histogram.


```python
n = input().lower().split()
vowel=['a','e','i','o','u']
max=0
count=0
j=[]
for word in n:
    if len(word)>max:
        max= len(word)
        k=word
    if word[0] in vowel:
        count+=1
    j.append(len(word))
    
print (f"Longest String is {k} with {max} characters")
print (count)
print(j)

import matplotlib.pyplot as plt
plt.hist(j,bins=range(min(j), max(j)))
plt.show()
    
```

## Sorting Dictionaries:
### Write a function to sort a list of dictionaries by a key.

Example:

data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
Sort by age → [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]



```python
data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
l = sorted(data, key=lambda x: x['age'])

print(l)

```
