## User Input and Basic Statistics:
### Write a program to read a comma-separated list of numbers from the user (e.g., “10, 20, 30, 40”).
### Calculate the mean of the numbers.
### Identify and print the smallest and largest numbers.
### Plot a histogram of the numbers.

```python
n= list(map (int,input().split(",")))
print (n)

## Converting- 10, 20, 30, 40 into a list which consists of integers
```

```python
import statistics
mean=statistics.mean(n)
print (mean)
```

```python
c= n[0]
d=n[0]
for word in n:
    if word < c:
        c=word
    if word > d:
        d=word
print (f"Smallest is ", c)
print (f"Largest is ",d)
    
```

```python
import matplotlib.pyplot as plt
plt.hist(n, bins=10, edgecolor='black')  # You can adjust the number of bins
plt.title("Histogram of Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
