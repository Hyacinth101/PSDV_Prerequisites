Q 14) Word filtering and SOrting
Write a program to:
Split a string into a list of words.
Remove all words that are shorter than 4 characters.
Sort the remaining words alphabetically.
Visualize the word lengths using a bar chart.


```python
import matplotlib.pyplot as plt
n= input()
word = n.split()
wordz = sorted([i for i in word if len(i) >= 4])

leng = [len(i) for i in wordz]


plt.bar(wordz, leng)
plt.xlabel('Sorted Words, with more than or equal to 4 character')
plt.ylabel('Length of Word of these sorted words')
plt.title('Q.14')
plt.xticks(rotation=90)
plt.show()

print(wordz)
```

## Q.15) Anagram FInder
Write a Python function to find all anagrams of a word in a given list.
Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
Output: ["enlist", "inlets"]

```python
main= "listen"
main=sorted (main)
n= ["enlist", "google", "inlets", "banana"]
anagram=[]
for i in n:
    sort= sorted(i)
    if sort == main:
        anagram.append(i)

print (anagram)
```
