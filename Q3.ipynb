### Fibonacci Sequence:
Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
Plot the Fibonacci sequence using a line graph.

```python
n=int(input())
a=0
b=1
fib=[a,b]
print (a, end=' ')
print (b, end=' ')
for i in range (0,n-2):
    c=a+b
    fib.append(c)
    print (c, end=' ')
    a=b
    b=c

    
```

#### Plotting the line graph (Using MatplotLib) 

```python


import matplotlib.pyplot as py
plt.plot(range(n), fib, marker='o')
plt.title("Fibonacci Sequence")
plt.xlabel("Index")
plt.ylabel("Fibonacci Number")
plt.grid(True)
plt.show()
```
